```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visual style
sns.set(style="whitegrid")

# Group by segment and calculate average interest
avg_interest = df.groupby("Community Segment")["Interest in Mushrooms"].mean().reset_index()

# Sort for clearer display
avg_interest = avg_interest.sort_values(by="Interest in Mushrooms", ascending=False)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(
    data=avg_interest,
    x="Interest in Mushrooms",
    y="Community Segment",
    palette="viridis"
)

# Labels and title
plt.title("Average Interest in Mushrooms by Community Segment", fontsize=14)
plt.xlabel("Average Interest (1–5 scale)")
plt.ylabel("Community Segment")
plt.xlim(0, 5)
plt.tight_layout()
plt.show()
```

```python
# Calculate the average of all metrics by group
heatmap_data = df.groupby("Community Segment").mean(numeric_only=True)

# Set up the plot
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, cmap="YlGnBu", linewidths=0.5, linecolor='gray')

# Add labels
plt.title("Community Needs Profile Heatmap", fontsize=14)
plt.xlabel("Program Factors")
plt.ylabel("Community Segment")
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Select features for clustering
features = df.drop(columns=["Community Segment"])

# Normalize features for clustering
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Apply KMeans clustering (3 groups for simplicity)
kmeans = KMeans(n_clusters=3, random_state=42)
df["Cluster"] = kmeans.fit_predict(scaled_features)

# Reduce dimensions for visualization
pca = PCA(n_components=2)
pca_components = pca.fit_transform(scaled_features)
df["PCA1"] = pca_components[:, 0]
df["PCA2"] = pca_components[:, 1]

# Plot the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x="PCA1",
    y="PCA2",
    hue="Cluster",
    style="Community Segment",
    palette="Set2",
    s=100
)
plt.title("Quantum-Inspired Clustering of Community Profiles", fontsize=14)
plt.xlabel("PCA Dimension 1")
plt.ylabel("PCA Dimension 2")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import random

# Define updated Texas-based community segments
texas_segments = [
    "South Dallas Residents",
    "Austin Gardeners & Foragers",
    "Central TX Wellness Seekers",
    "Latino/Indigenous Herbalists"
]

# Initialize dictionary for synthetic data
new_data = {
    "Community Segment": [],
    "Interest in Mushrooms": [],
    "Interest in Plants": [],
    "Access to Resources": [],
    "Perceived Health Benefit": [],
    "Perceived Cost Barrier": [],
    "Willingness to Participate": []
}

# Simulate data per group
for segment in texas_segments:
    for _ in range(25):
        new_data["Community Segment"].append(segment)

        if segment == "South Dallas Residents":
            new_data["Interest in Mushrooms"].append(random.randint(2, 4))
            new_data["Interest in Plants"].append(random.randint(2, 4))
            new_data["Access to Resources"].append(random.randint(1, 3))
            new_data["Perceived Health Benefit"].append(random.randint(3, 5))
            new_data["Perceived Cost Barrier"].append(random.randint(3, 4))
            new_data["Willingness to Participate"].append(random.randint(2, 4))

        elif segment == "Austin Gardeners & Foragers":
            new_data["Interest in Mushrooms"].append(random.randint(4, 5))
            new_data["Interest in Plants"].append(random.randint(4, 5))
            new_data["Access to Resources"].append(random.randint(4, 5))
            new_data["Perceived Health Benefit"].append(random.randint(4, 5))
            new_data["Perceived Cost Barrier"].append(random.randint(1, 2))
            new_data["Willingness to Participate"].append(random.randint(4, 5))

        elif segment == "Central TX Wellness Seekers":
            new_data["Interest in Mushrooms"].append(random.randint(3, 5))
            new_data["Interest in Plants"].append(random.randint(3, 5))
            new_data["Access to Resources"].append(random.randint(3, 4))
            new_data["Perceived Health Benefit"].append(random.randint(4, 5))
            new_data["Perceived Cost Barrier"].append(random.randint(2, 3))
            new_data["Willingness to Participate"].append(random.randint(3, 5))

        elif segment == "Latino/Indigenous Herbalists":
            new_data["Interest in Mushrooms"].append(random.randint(3, 5))
            new_data["Interest in Plants"].append(random.randint(4, 5))
            new_data["Access to Resources"].append(random.randint(2, 4))
            new_data["Perceived Health Benefit"].append(random.randint(4, 5))
            new_data["Perceived Cost Barrier"].append(random.randint(2, 3))
            new_data["Willingness to Participate"].append(random.randint(3, 5))

# Convert to DataFrame
df_tx = pd.DataFrame(new_data)

# Preview
df_tx.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the average of all metrics by group
heatmap_data_tx = df_tx.groupby("Community Segment").mean(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(
    heatmap_data_tx,
    annot=True,
    cmap="YlGnBu",
    linewidths=0.5,
    linecolor='gray'
)

# Add labels and formatting
plt.title("Community Needs Profile Heatmap (Texas-Based Segments)", fontsize=14)
plt.xlabel("Program Factors")
plt.ylabel("Community Segment")
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numeric features
features = df_tx.drop(columns=["Community Segment"])

# Normalize features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Run KMeans clustering (3 clusters)
kmeans = KMeans(n_clusters=3, random_state=42)
df_tx["Cluster"] = kmeans.fit_predict(scaled_features)

# Reduce to 2D using PCA for visualization
pca = PCA(n_components=2)
pca_components = pca.fit_transform(scaled_features)
df_tx["PCA1"] = pca_components[:, 0]
df_tx["PCA2"] = pca_components[:, 1]

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_tx,
    x="PCA1",
    y="PCA2",
    hue="Cluster",
    style="Community Segment",
    palette="Set2",
    s=100
)
plt.title("Quantum-Inspired Clustering of Texas Community Profiles", fontsize=14)
plt.xlabel("PCA Dimension 1")
plt.ylabel("PCA Dimension 2")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Define a function to assign recommended program type
def recommend_program(row):
    if row["Interest in Mushrooms"] >= 4 and row["Access to Resources"] <= 2:
        return "Intro Mushroom Workshop"
    elif row["Willingness to Participate"] >= 4 and row["Perceived Cost Barrier"] <= 2:
        return "Ambassador Leadership Track"
    elif row["Interest in Plants"] >= 4 and row["Access to Resources"] >= 4:
        return "Garden-to-Market Incubator"
    elif row["Perceived Cost Barrier"] >= 3 and row["Willingness to Participate"] >= 3:
        return "Scholarship Program Option"
    else:
        return "General Interest Program"

# Apply function to dataframe
df_tx["Recommended Program"] = df_tx.apply(recommend_program, axis=1)

# View sample output
df_tx[["Community Segment", "Interest in Mushrooms", "Access to Resources", "Willingness to Participate", "Perceived Cost Barrier", "Recommended Program"]].head(10)
```
